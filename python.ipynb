## Qu'est-ce qu'un ordinateur fait réellement ?

Un ordinateur est une machine **stupide mais rapide**. Il ne comprend rien, il n'a aucune intelligence. La seule chose qu'il sait faire, c'est suivre des instructions à la lettre, une par une, dans l'ordre où vous les lui donnez.

**Imaginez un robot cuisinier :**
- Ce robot ne sait pas ce qu'est un gâteau
- Il ne sait pas ce que "cuisiner" veut dire
- Il ne peut pas improviser
- MAIS il peut suivre exactement les instructions que vous lui donnez

Si vous lui dites :
1. Prends un bol
2. Mets 200g de farine dedans
3. Ajoute 2 œufs
4. Mélange pendant 2 minutes

Il le fera parfaitement. Mais si vous oubliez de dire "prends un bol", il essaiera de mettre la farine... nulle part, et ça plantera.

**C'est exactement ça, la programmation** : écrire des instructions ultra-précises pour une machine qui ne comprend rien mais exécute tout à la lettre.

## Pourquoi avons-nous besoin d'un langage de programmation ?

Le problème, c'est que l'ordinateur ne comprend qu'un seul langage : le **binaire** (des suites de 0 et de 1).

Par exemple, pour lui dire "additionne 2 et 3", il faudrait écrire quelque chose comme :
```
10110000 00000010 00000011
```

C'est illisible et impossible à écrire pour un humain !

**Solution** : On a inventé des **langages de programmation**. Ce sont des langages intermédiaires :
- Assez simples pour qu'un humain puisse les écrire et les comprendre
- Qui peuvent être traduits automatiquement en binaire pour l'ordinateur

**Python** est l'un de ces langages, et c'est l'un des plus simples à apprendre.

## Qu'est-ce que Python ?

Python est un **langage de programmation** créé en 1991 par Guido Van Rossum (un informaticien néerlandais). 

**Pourquoi le nom "Python" ?**
Ce n'est pas à cause du serpent ! Guido était fan d'une série comique britannique appelée "Monty Python's Flying Circus", et il a nommé son langage en hommage à cette série.

**Pourquoi Python est-il si populaire ?**

1. **Il est facile à lire** : Le code Python ressemble presque à de l'anglais. Même sans connaître Python, vous pouvez souvent deviner ce que fait un programme.

2. **Il est permissif** : Python vous laisse faire beaucoup de choses sans vous compliquer la vie avec des règles strictes.

3. **Il est polyvalent** : On peut faire presque tout avec Python :
   - Des sites web
   - De l'intelligence artificielle
   - Des jeux vidéo
   - De l'analyse de données
   - De l'automatisation de tâches répétitives
   - Des applications de bureau
   - Et bien plus...

4. **Il a une énorme communauté** : Des millions de personnes utilisent Python. Si vous avez un problème, quelqu'un a probablement déjà trouvé la solution.

5. **Il est gratuit** : Python est open source, tout le monde peut l'utiliser gratuitement.

## Comment Python exécute votre code ?

Quand vous écrivez du code Python, voici ce qui se passe :

```
Vous écrivez du code Python (lisible par un humain)
            ↓
L'interpréteur Python lit votre code ligne par ligne
            ↓
Il traduit chaque ligne en instructions que l'ordinateur comprend
            ↓
L'ordinateur exécute ces instructions
            ↓
Vous voyez le résultat
```

**L'interpréteur** est le programme qui fait cette traduction. C'est lui qui "comprend" Python et le traduit pour l'ordinateur.

**Point important** : Python est un langage **interprété**, ce qui veut dire qu'il traduit et exécute votre code **ligne par ligne**. Si une ligne a une erreur, Python s'arrête à cette ligne et vous dit qu'il y a un problème.

### La multiplication : `*`

**À quoi ça sert ?** Multiplier deux nombres.

**Attention !** En programmation, on utilise `*` (l'astérisque) et non `×`. Pourquoi ? Parce que `×` n'existe pas sur un clavier standard !

### La division : `/`

**À quoi ça sert ?** Diviser un nombre par un autre.

````python
10 / 2
````

````python
15 / 4
````

````python
7 / 3
````

**Observation importante** : Remarquez que `10 / 2` donne `5.0` et non `5`. Le `.0` signifie que c'est un nombre décimal (avec virgule). En Python 3, la division donne TOUJOURS un nombre décimal, même si le résultat est "rond".

**Pourquoi ?** Parce que dans la vraie vie, une division peut donner n'importe quel résultat. Python préfère toujours vous donner un nombre décimal pour être cohérent.

### La division entière : `//`

**À quoi ça sert ?** Diviser et ne garder que la partie entière (avant la virgule).

````python
# Division normale
10 / 3
````

````python
# Division entière : on jette tout ce qui est après la virgule
10 // 3
````

**Comparaison :**
- `10 / 3 = 3.33333...` (division normale)
- `10 // 3 = 3` (on jette tout ce qui est après la virgule)

**Quand l'utiliser ?** Quand vous voulez savoir "combien de fois un nombre rentre dans un autre" :
- Vous avez 17 bonbons à partager entre 5 enfants
- `17 // 5 = 3` → Chaque enfant aura 3 bonbons

````python
# Combien de bonbons par enfant ?
17 // 5
````

### Le modulo : `%`

**À quoi ça sert ?** Trouver le **reste** d'une division.

C'est peut-être l'opérateur le plus mystérieux pour les débutants, alors expliquons bien.

**Exemple concret** : Vous avez 17 bonbons à partager entre 5 enfants.
- Chaque enfant reçoit 3 bonbons (car `17 // 5 = 3`)
- Vous avez distribué `3 × 5 = 15` bonbons
- Il vous reste `17 - 15 = 2` bonbons

Ce "reste", c'est ce que calcule le modulo :

````python
# Combien de bonbons restent après le partage ?
17 % 5
````

````python
# Autres exemples
print("10 % 3 =", 10 % 3)  # 10 = 3×3 + 1, le reste est 1
print("20 % 4 =", 20 % 4)  # 20 = 4×5 + 0, le reste est 0 (division exacte)
print("7 % 2 =", 7 % 2)    # 7 = 2×3 + 1, le reste est 1
````

**Utilisation super pratique** : Savoir si un nombre est pair ou impair !
- Un nombre est pair si le reste de sa division par 2 est 0
- Un nombre est impair si le reste de sa division par 2 est 1

````python
# 10 est-il pair ?
print("10 % 2 =", 10 % 2)  # 0 → 10 est pair
````

````python
# 7 est-il pair ?
print("7 % 2 =", 7 % 2)    # 1 → 7 est impair
````

### La puissance : `**`

**À quoi ça sert ?** Élever un nombre à une puissance (exposant).

**Rappel mathématique** :
- 2³ = 2 × 2 × 2 = 8
- 5² = 5 × 5 = 25

En Python, on écrit `2 ** 3` au lieu de 2³ :

````python
# 2 puissance 3 = 2 × 2 × 2
2 ** 3
````

````python
# 5 au carré = 5 × 5
5 ** 2
````

````python
# 10 puissance 6 = 1 million
10 ** 6
````

**Pourquoi `**` ?** Parce qu'on ne peut pas écrire des exposants en hauteur sur un clavier. Le double astérisque est une convention adoptée par de nombreux langages.

### Priorité des opérations : Python respecte les maths !

En mathématiques, il y a un ordre de priorité :
1. Les parenthèses d'abord
2. Puis les puissances
3. Puis les multiplications et divisions
4. Enfin les additions et soustractions

Python respecte exactement ces règles :

````python
# Python fait d'abord 3×4=12, puis 2+12=14
2 + 3 * 4
````

````python
# Les parenthèses forcent Python à faire d'abord 2+3=5, puis 5×4=20
(2 + 3) * 4
````

**Conseil** : En cas de doute, mettez des parenthèses ! Ça rend aussi le code plus lisible.

### Les espaces n'ont pas d'importance (pour les calculs)

Python ignore les espaces autour des opérateurs. Tous ces exemples sont équivalents :

````python
print(2+3)
print(2 + 3)
print(2  +  3)
print(   2+3)
````

**Conseil de style** : Par convention, on met un espace avant et après chaque opérateur pour une meilleure lisibilité : `2 + 3` plutôt que `2+3`.

## Pourquoi les "types" existent-ils ?

Dans la vraie vie, nous manipulons différentes sortes de choses :
- Des nombres (votre âge, le prix d'un article...)
- Du texte (votre nom, une adresse...)
- Des réponses oui/non (Êtes-vous majeur ? Le magasin est-il ouvert ?...)

L'ordinateur a besoin de savoir **quel type de chose** il manipule pour savoir **quoi en faire**.

**Exemple de la vraie vie :**
- Si je vous dis "ajoutez 5 et 3", vous faites 5 + 3 = 8 (c'est un calcul)
- Si je vous dis "ajoutez 'Bon' et 'jour'", vous faites "Bonjour" (vous collez les mots)

Vous savez instinctivement faire la différence. L'ordinateur, lui, a besoin qu'on lui dise explicitement quel type de donnée il manipule.

## Type 1 : Les nombres entiers (`int`)

### Qu'est-ce que c'est ?

Un **entier** (en anglais "integer", abrégé `int`) est un nombre sans virgule :
- Positif : 1, 2, 42, 1000000
- Négatif : -1, -50, -999
- Zéro : 0

### À quoi ça sert ?

Les entiers servent à compter des choses qu'on ne peut pas diviser :
- Le nombre de personnes dans une pièce (on ne peut pas avoir 3.5 personnes)
- L'âge de quelqu'un en années (25 ans, pas 25.7 ans)
- Le nombre de produits dans un panier
- Un numéro de téléphone, un code postal...

````python
# Exemples d'entiers
print(42)
print(-17)
print(0)
print(1000000000)
````

### Comment savoir si quelque chose est un entier ?

Python a une fonction `type()` qui vous dit le type d'une donnée :

````python
type(42)
````

Le mot `int` confirme que 42 est un entier.

**Qu'est-ce que ça veut dire `<class 'int'>` ?**
Pour l'instant, retenez simplement que `int` = entier. Nous expliquerons le mot "class" bien plus tard.

## Type 2 : Les nombres décimaux (`float`)

### Qu'est-ce que c'est ?

Un **flottant** (en anglais "floating-point number", abrégé `float`) est un nombre avec une partie décimale :
- 3.14
- -0.5
- 2.0 (même si la partie décimale est 0, c'est un float !)

### ATTENTION : Le point, pas la virgule !

En programmation, on utilise le **point** comme séparateur décimal, pas la virgule :
- ✅ Correct : `3.14`
- ❌ Incorrect : `3,14` (Python ne comprendra pas)

**Pourquoi ?** Les langages de programmation sont basés sur l'anglais, et en anglais, on utilise le point pour les décimaux.

### À quoi ça sert ?

Les décimaux servent pour tout ce qui peut être "partiel" :
- Un prix : 19.99 €
- Une température : 36.7°C
- Une distance : 42.195 km
- Un pourcentage : 0.75 (75%)

````python
# Exemples de flottants
print(3.14)
print(-0.001)
print(2.0)  # Même si c'est "2", le point en fait un décimal
````

````python
# Vérifier le type
type(3.14)
````

### La division crée toujours des floats

Même si le résultat est "rond", la division `/` donne un float :

````python
print(10 / 2)
print(type(10 / 2))
````

**Pourquoi ?** C'est un choix de conception de Python 3. Comme une division peut donner n'importe quel résultat, Python décide que le résultat est toujours un float pour être cohérent.

### Attention aux calculs avec des floats !

Les ordinateurs ont du mal à représenter exactement les nombres décimaux. Cela peut créer de petites erreurs :

````python
# On s'attend à 0.3, mais...
0.1 + 0.1 + 0.1
````

Vous vous attendiez à `0.3`, mais vous obtenez `0.30000000000000004`. C'est une limitation technique de tous les ordinateurs, pas un bug de Python.

**Pour l'instant**, ne vous inquiétez pas de ça. Sachez juste que ça existe.

## Type 3 : Les chaînes de caractères (`str`)

### Qu'est-ce que c'est ?

Une **chaîne de caractères** (en anglais "string", abrégé `str`) est du texte. 

Le mot "chaîne" vient de l'idée que c'est une "chaîne" de caractères mis bout à bout, comme les maillons d'une chaîne.

### Comment écrire une chaîne ?

On entoure le texte de guillemets. Vous avez deux choix :
- Guillemets simples : `'Bonjour'`
- Guillemets doubles : `"Bonjour"`

Les deux sont exactement équivalents. Choisissez celui que vous préférez.

````python
print('Bonjour')
print("Bonjour")
print(type('Bonjour'))
````

### Pourquoi les guillemets sont obligatoires ?

Pour que Python fasse la différence entre du texte et des instructions.

````python
# Sans guillemets : Python pense que c'est le nom d'une variable
# Décommentez la ligne suivante pour voir l'erreur :
# Bonjour
````

````python
# Avec guillemets : Python comprend que c'est du texte brut
'Bonjour'
````

### Attention : `'42'` n'est pas pareil que `42` !

C'est très important de comprendre cette différence :

````python
# 42 sans guillemets : c'est un NOMBRE
print(42)
print(type(42))
````

````python
# '42' avec guillemets : c'est du TEXTE
print('42')
print(type('42'))
````

- `42` (sans guillemets) est le nombre quarante-deux. On peut faire des calculs avec.
- `'42'` (avec guillemets) est le texte "42", comme si vous aviez écrit "quarante-deux". On ne peut PAS faire de calculs mathématiques avec.

**Analogie :** C'est comme la différence entre :
- Le nombre 7 sur une calculatrice (vous pouvez calculer avec)
- Le chiffre "7" écrit sur un bout de papier (c'est juste un dessin)

## Type 4 : Les booléens (`bool`)

### Qu'est-ce que c'est ?

Un **booléen** (en anglais "boolean", abrégé `bool`) est une valeur qui ne peut être que :
- `True` (Vrai)
- `False` (Faux)

C'est tout. Il n'y a pas d'autre possibilité.

**Attention à la majuscule !** C'est `True` avec un T majuscule, pas `true`.

````python
print(True)
print(False)
print(type(True))
````

### Pourquoi le nom "booléen" ?

En hommage au mathématicien George Boole (1815-1864) qui a inventé l'algèbre booléenne, la base de toute la logique informatique.

### À quoi ça sert ?

Les booléens servent à répondre à des questions oui/non :
- L'utilisateur est-il connecté ? `True` ou `False`
- Le panier est-il vide ? `True` ou `False`
- La température est-elle supérieure à 30° ? `True` ou `False`

Ils sont **essentiels** pour prendre des décisions dans un programme (on verra ça avec les conditions).

### Comment obtenir un booléen ? Les comparaisons

On obtient des booléens en comparant des valeurs :

````python
# 5 est-il plus grand que 3 ?
5 > 3
````

````python
# 5 est-il plus petit que 3 ?
5 < 3
````

````python
# 5 est-il égal à 5 ?
5 == 5
````

````python
# 5 est-il égal à 3 ?
5 == 3
````

**Liste des opérateurs de comparaison :**

| Opérateur | Signification | Exemple | Résultat |
|-----------|---------------|---------|----------|
| `==` | Est égal à | `5 == 5` | `True` |
| `!=` | Est différent de | `5 != 3` | `True` |
| `<` | Est inférieur à | `3 < 5` | `True` |
| `>` | Est supérieur à | `5 > 3` | `True` |
| `<=` | Est inférieur ou égal à | `5 <= 5` | `True` |
| `>=` | Est supérieur ou égal à | `5 >= 3` | `True` |

### ⚠️ ATTENTION : `=` et `==` sont DIFFÉRENTS !

C'est l'erreur la plus courante chez les débutants :

- `=` sert à **assigner** une valeur (mettre quelque chose dans une boîte)
- `==` sert à **comparer** deux valeurs (demander "sont-ils égaux ?")

````python
# = c'est l'assignation (on verra ça avec les variables)
age = 25        # "Je mets 25 dans la boîte 'age'"
print(age)
````

````python
# == c'est la comparaison
age == 25       # "Est-ce que age vaut 25 ?" → True ou False
````

# CHAPITRE 3 : LES VARIABLES - DONNER DES NOMS AUX CHOSES
---

## Le problème : comment se souvenir des valeurs ?

Jusqu'ici, nous avons fait des calculs simples :

````python
2 + 3
````

Mais le résultat `5` est affiché puis... perdu ! Si vous voulez l'utiliser plus tard, vous devez refaire le calcul.

**Imaginez** : Vous calculez le prix total d'un panier (47.50 €), puis vous voulez calculer les taxes dessus, puis appliquer une réduction... Vous devriez retaper 47.50 à chaque fois, et si le prix change, vous devriez tout refaire !

**Solution** : Les **variables** permettent de stocker des valeurs pour les réutiliser.

## Qu'est-ce qu'une variable ?

Une **variable** est une "boîte" avec un nom dans laquelle on range une valeur.

**Analogie détaillée :**

Imaginez un entrepôt plein de boîtes. Chaque boîte a :
- Une **étiquette** (le nom de la variable)
- Un **contenu** (la valeur stockée)

Quand vous créez une variable :
1. Vous créez une nouvelle boîte
2. Vous collez une étiquette dessus (le nom)
3. Vous mettez quelque chose dedans (la valeur)

Plus tard, vous pouvez :
- Regarder ce qu'il y a dans la boîte (lire la variable)
- Changer le contenu de la boîte (modifier la variable)
- Utiliser le contenu dans un calcul

## Comment créer une variable ?

On utilise le signe `=` (appelé **opérateur d'affectation**) :

````python
age = 25
````

**Ce qui se passe :**
1. Python crée une "boîte" en mémoire
2. Python colle l'étiquette `age` sur cette boîte
3. Python met la valeur `25` dans cette boîte

**Visualisation :**
```
┌─────────────┐
│     25      │  ← La valeur
├─────────────┤
│    age      │  ← Le nom (l'étiquette)
└─────────────┘
```

## Comment utiliser une variable ?

Une fois créée, vous pouvez utiliser le **nom** de la variable partout où vous auriez mis la **valeur** :

````python
age = 25
print(age)        # Affiche 25
````

````python
age + 10          # Calcule 25 + 10
````

````python
age * 2           # Calcule 25 * 2
````

Quand Python voit `age`, il va chercher dans sa mémoire : "Qu'est-ce qu'il y a dans la boîte étiquetée 'age' ?" et trouve `25`.

## Pourquoi c'est utile ? Exemples concrets

### Exemple 1 : Calculer une facture

**Sans variables (mauvais) :**

````python
# Sans variables - Si le prix change, il faut tout refaire !
print("TVA:", 47.50 * 0.20)
print("Prix TTC:", 47.50 + 9.5)
print("Après réduction 10%:", 57.0 * 0.90)
````

**Avec variables (bien) :**

````python
# Avec variables - Si le prix change, on modifie juste prix_ht
prix_ht = 47.50
tva = prix_ht * 0.20
prix_ttc = prix_ht + tva
prix_final = prix_ttc * 0.90

print("Prix HT:", prix_ht)
print("TVA:", tva)
print("Prix TTC:", prix_ttc)
print("Prix final (après 10% de réduction):", prix_final)
````

### Exemple 2 : Informations d'un utilisateur

````python
nom = 'Marie'
age = 28
ville = 'Paris'

print(nom, "a", age, "ans et habite à", ville)
````

## Règles pour nommer une variable

Vous ne pouvez pas appeler une variable n'importe comment. Voici les règles :

### Règle 1 : Caractères autorisés

Une variable peut contenir :
- Des lettres (a-z, A-Z)
- Des chiffres (0-9)
- Le caractère souligné `_` (underscore)

**C'est tout !** Pas d'espaces, pas de tirets, pas de caractères spéciaux.

### Règle 2 : Ne peut pas commencer par un chiffre

````python
# ✅ Correct
age = 25
age2 = 30
compteur_1 = 0

# ❌ Incorrect - décommentez pour voir l'erreur
# 2age = 25         # Commence par un chiffre → ERREUR
````

## Modifier une variable

On peut changer la valeur d'une variable à tout moment en lui assignant une nouvelle valeur :

### Le cas courant : incrémenter

Très souvent, on veut augmenter une variable de 1 (par exemple, un compteur) :

````python
compteur = 0
print("Début: compteur =", compteur)

compteur = compteur + 1
print("Après +1: compteur =", compteur)

compteur = compteur + 1
print("Après +1: compteur =", compteur)
````

**Comment lire `compteur = compteur + 1` ?**

⚠️ Ce n'est PAS une équation mathématique ! En maths, `x = x + 1` n'a pas de sens.

En programmation, c'est une **instruction** qui se lit de droite à gauche :
1. `compteur + 1` → Prendre la valeur actuelle de compteur (0) et ajouter 1 → 1
2. `compteur =` → Mettre ce résultat (1) dans la variable compteur

**C'est comme dire :** "La nouvelle valeur de compteur est égale à l'ancienne valeur de compteur plus 1."

### Raccourci : les opérateurs combinés

Comme `variable = variable + quelque_chose` est très courant, Python propose des raccourcis :

| Forme longue | Raccourci | Exemple |
|--------------|-----------|---------||
| `x = x + 5` | `x += 5` | Ajouter 5 à x |
| `x = x - 5` | `x -= 5` | Soustraire 5 de x |
| `x = x * 5` | `x *= 5` | Multiplier x par 5 |
| `x = x / 5` | `x /= 5` | Diviser x par 5 |
| `x = x ** 2` | `x **= 2` | x au carré |

````python
score = 100
print("Début: score =", score)

score += 10      # Équivalent à : score = score + 10
print("Après += 10: score =", score)

score -= 20      # Équivalent à : score = score - 20
print("Après -= 20: score =", score)

score *= 2       # Équivalent à : score = score * 2
print("Après *= 2: score =", score)
````

## Pourquoi les conditions existent-elles ?

Jusqu'ici, nos programmes exécutaient TOUTES les instructions, dans l'ordre. Mais dans la vraie vie, on veut souvent que le programme fasse des choix :

- **SI** l'utilisateur a plus de 18 ans, **ALORS** afficher "Vous êtes majeur"
- **SI** le mot de passe est correct, **ALORS** connecter l'utilisateur, **SINON** afficher une erreur
- **SI** le panier est vide, **ALORS** afficher "Votre panier est vide"

Les **conditions** permettent au programme de prendre des décisions.

## Le bloc `if` : "Si..."

### Structure de base

```python
if condition:
    # Code exécuté si la condition est vraie
    instruction1
    instruction2
```

**Les éléments :**
1. Le mot-clé `if`
2. Une **condition** (quelque chose qui est `True` ou `False`)
3. Les deux-points `:` (OBLIGATOIRES !)
4. Un bloc de code **indenté** (décalé de 4 espaces)

### Comment ça fonctionne ?

```
      ┌──────────────┐
      │  condition   │
      │  vraie ?     │
      └──────┬───────┘
             │
      ┌──────┴──────┐
      │             │
    Oui           Non
      │             │
      ▼             │
┌───────────┐       │
│ Exécuter  │       │
│ le bloc   │       │
└─────┬─────┘       │
      │             │
      └──────┬──────┘
             │
             ▼
      Suite du programme
```

### Exemple simple

````python
age = 20

if age >= 18:
    print('Vous êtes majeur.')

print('Fin du programme.')
````

**Exécution pas à pas :**
1. `age` reçoit la valeur `20`
2. Python évalue `age >= 18` → `20 >= 18` → `True`
3. La condition est vraie, donc Python exécute le bloc indenté
4. Affiche "Vous êtes majeur."
5. Fin du bloc `if` (l'indentation diminue)
6. Affiche "Fin du programme."

````python
# Testons avec age = 15
age = 15

if age >= 18:
    print('Vous êtes majeur.')  # Cette ligne est sautée !

print('Fin du programme.')  # Cette ligne s'exécute toujours
````

**Tout ce qui est indenté fait partie du `if`** :

````python
age = 20

if age >= 18:
    print('Vous êtes majeur.')           # DANS le if
    print('Vous pouvez voter.')          # DANS le if
    print('Vous pouvez conduire.')       # DANS le if

print('Au revoir.')                       # HORS du if (pas indenté)
````

````python
# Si age = 15, les trois lignes indentées sont sautées
age = 15

if age >= 18:
    print('Vous êtes majeur.')           # SAUTÉ
    print('Vous pouvez voter.')          # SAUTÉ
    print('Vous pouvez conduire.')       # SAUTÉ

print('Au revoir.')                       # AFFICHÉ (pas dans le if)
````

## Le bloc `if-else` : "Si... sinon..."

### Pourquoi ?

Souvent, on veut faire quelque chose si une condition est vraie, et **autre chose** si elle est fausse.

````python
age = 15

if age >= 18:
    print('Vous êtes majeur.')
else:
    print('Vous êtes mineur.')
````

**Exécution :**
1. `age >= 18` → `15 >= 18` → `False`
2. La condition est fausse, donc on exécute le bloc `else`
3. Affiche "Vous êtes mineur."

### Structure

```python
if condition:
    # Exécuté si la condition est vraie
else:
    # Exécuté si la condition est fausse
```

**Important :** `else` n'a PAS de condition. Il attrape tout ce qui n'est pas géré par le `if`.

## Le bloc `if-elif-else` : "Si... sinon si... sinon..."

### Pourquoi ?

Parfois, il y a plus de deux possibilités :

- Si la note est >= 16 : "Très bien"
- Sinon, si la note est >= 14 : "Bien"
- Sinon, si la note est >= 12 : "Assez bien"
- Sinon, si la note est >= 10 : "Passable"
- Sinon : "Insuffisant"

````python
note = 14

if note >= 16:
    print('Très bien')
elif note >= 14:
    print('Bien')
elif note >= 12:
    print('Assez bien')
elif note >= 10:
    print('Passable')
else:
    print('Insuffisant')
````

**`elif`** est la contraction de "else if" (sinon si).

### Comment ça fonctionne ?

Python teste les conditions **dans l'ordre**, de haut en bas :

1. `note >= 16` → `14 >= 16` → `False` → On passe à la suite
2. `note >= 14` → `14 >= 14` → `True` → On exécute ce bloc !
3. **Les autres conditions ne sont même pas testées**
4. Affiche "Bien"

### ⚠️ L'ORDRE DES CONDITIONS EST CRUCIAL !

**MAUVAIS exemple :**

````python
# MAUVAIS CODE - l'ordre est incorrect !
note = 18

if note >= 10:           # 18 >= 10 → True → On s'arrête ici !
    print('Passable')    # Affiche "Passable" alors que 18 mérite "Très bien" !
elif note >= 14:
    print('Bien')        # Jamais atteint
elif note >= 16:
    print('Très bien')   # Jamais atteint !
````

**Pourquoi ?** `18 >= 10` est vrai, donc Python s'arrête là.

**RÈGLE :** Mettez les conditions les plus restrictives EN PREMIER.

````python
# BON CODE - l'ordre est correct !
note = 18

if note >= 16:           # Condition la plus restrictive en premier
    print('Très bien')
elif note >= 14:
    print('Bien')
elif note >= 10:
    print('Passable')
else:
    print('Insuffisant')
````

````python
# <= : Est inférieur ou égal à
print("3 <= 5 :", 3 <= 5)
print("5 <= 5 :", 5 <= 5)  # 5 est égal à 5, donc c'est vrai
print("6 <= 5 :", 6 <= 5)
````

````python
# >= : Est supérieur ou égal à
print("5 >= 3 :", 5 >= 3)
print("5 >= 5 :", 5 >= 5)  # 5 est égal à 5, donc c'est vrai
print("4 >= 5 :", 4 >= 5)
````

## Combiner les conditions : `and`, `or`, `not`

### `and` : ET logique

La condition complète est vraie **seulement si les deux parties sont vraies**.

| A | B | A and B |
|---|---|---------||
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

````python
# Exemple : Pour conduire, il faut avoir 18 ans ET avoir le permis
age = 25
permis = True

if age >= 18 and permis:
    print('Vous pouvez conduire.')
else:
    print('Vous ne pouvez pas conduire.')
````

````python
# Testons avec quelqu'un qui a 25 ans mais pas de permis
age = 25
permis = False

if age >= 18 and permis:
    print('Vous pouvez conduire.')
else:
    print('Vous ne pouvez pas conduire.')  # Les deux conditions ne sont pas vraies
````

### `or` : OU logique

La condition complète est vraie **si au moins une des parties est vraie**.

| A | B | A or B |
|---|---|--------|
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

````python
# Exemple : C'est le week-end si c'est samedi OU dimanche
jour = 'samedi'

if jour == 'samedi' or jour == 'dimanche':
    print('C\'est le week-end !')
else:
    print('C\'est un jour de semaine.')
````

### `not` : NON logique (inverse)

Inverse la valeur : `True` devient `False`, `False` devient `True`.

| A | not A |
|---|-------|
| True | False |
| False | True |

````python
# Exemple : Afficher un message si l'utilisateur N'EST PAS connecté
connecte = False

if not connecte:
    print('Vous n\'êtes pas connecté. Veuillez vous connecter.')
````

### Combiner plusieurs opérateurs

````python
age = 25
abonnement = True
jour = 'dimanche'

# Entrée gratuite si (majeur ET abonné) OU si c'est dimanche
if (age >= 18 and abonnement) or jour == 'dimanche':
    print('Entrée gratuite !')
else:
    print('Entrée payante.')
````

**Conseil :** Utilisez des parenthèses pour clarifier, même si ce n'est pas toujours obligatoire.

## Pourquoi les boucles existent-elles ?

Imaginez que vous voulez afficher les nombres de 1 à 10 :

**Sans boucle :**

````python
# Sans boucle - beaucoup de répétition !
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
````

Et si c'était de 1 à 1000 ? 10000 ?

**Les boucles** permettent de répéter des instructions automatiquement.

## La boucle `while` : "Tant que..."

### Qu'est-ce que c'est ?

`while` répète un bloc de code **tant qu'une condition est vraie**.

### Structure

```python
while condition:
    # Code répété tant que la condition est vraie
    instruction1
    instruction2
```

### Exemple simple

````python
compteur = 1

while compteur <= 5:
    print(compteur)
    compteur = compteur + 1

print('Fini !')
````

**Exécution pas à pas :**

| Étape | compteur | compteur <= 5 ? | Action |
|-------|----------|-----------------|--------|
| 1 | 1 | True | Affiche 1, compteur devient 2 |
| 2 | 2 | True | Affiche 2, compteur devient 3 |
| 3 | 3 | True | Affiche 3, compteur devient 4 |
| 4 | 4 | True | Affiche 4, compteur devient 5 |
| 5 | 5 | True | Affiche 5, compteur devient 6 |
| 6 | 6 | False | Fin de la boucle |

### ⚠️ DANGER : La boucle infinie !

**Si la condition reste toujours vraie, la boucle ne s'arrête JAMAIS :**

```python
# MAUVAIS CODE - NE PAS EXÉCUTER !
compteur = 1

while compteur <= 5:
    print(compteur)
    # Oups ! On a oublié de modifier compteur !
```

`compteur` reste toujours à 1, donc `compteur <= 5` est toujours vrai. Le programme affiche `1` à l'infini jusqu'à ce que vous l'arrêtiez (Ctrl+C).

**Règle d'or :** Dans une boucle `while`, il FAUT que quelque chose change pour que la condition puisse devenir fausse un jour.

## La boucle `for` : "Pour chaque..."

### Qu'est-ce que c'est ?

`for` parcourt les éléments d'une séquence (une liste, une chaîne, un range...), un par un.

### Structure

```python
for element in sequence:
    # Code exécuté pour chaque élément
    # 'element' prend successivement chaque valeur de la séquence
```

### La fonction `range()` : Générer une séquence de nombres

`range()` crée une séquence de nombres. C'est souvent utilisé avec `for`.

**Syntaxes :**

```python
range(fin)              # De 0 à fin-1
range(debut, fin)       # De debut à fin-1
range(debut, fin, pas)  # De debut à fin-1, en avançant de 'pas'
```

````python
# Exemples de range()
print("range(5):", list(range(5)))              # De 0 à 4 (5 est exclu)
print("range(1, 6):", list(range(1, 6)))        # De 1 à 5 (6 est exclu)
print("range(0, 10, 2):", list(range(0, 10, 2))) # De 0 à 9, de 2 en 2
print("range(5, 0, -1):", list(range(5, 0, -1))) # Compte à rebours
````

**Pourquoi la fin est-elle exclue ?**

C'est une convention en informatique. `range(5)` donne 5 éléments : 0, 1, 2, 3, 4.

### Exemple avec `for` et `range`

````python
for i in range(5):
    print(i)
````

**Explication :** À chaque tour de boucle, `i` prend la valeur suivante de `range(5)`.

````python
# Afficher les nombres de 1 à 10
for i in range(1, 11):
    print(i)
````

**Explication :** À chaque tour, `lettre` prend le caractère suivant de la chaîne.

### Exemples pratiques

````python
# Compter les voyelles dans un mot
mot = 'Programmation'
compteur = 0
voyelles = 'aeiouAEIOU'

for lettre in mot:
    if lettre in voyelles:
        compteur += 1

print(f'Il y a {compteur} voyelles dans "{mot}"')
````

## Contrôler les boucles : `break` et `continue`

### `break` : Sortir immédiatement de la boucle

````python
for i in range(10):
    if i == 5:
        break           # Sort de la boucle immédiatement
    print(i)

print('Boucle terminée')
````

Quand `i` vaut 5, `break` interrompt la boucle. On n'affiche pas 5, 6, 7, 8, 9.

**À quoi ça sert ?** Arrêter la boucle quand on a trouvé ce qu'on cherchait, par exemple.

### `continue` : Passer à l'itération suivante

````python
for i in range(5):
    if i == 2:
        continue        # Saute le reste et passe au tour suivant
    print(i)
````

Quand `i` vaut 2, `continue` saute le `print(i)` et passe directement à `i = 3`.

**À quoi ça sert ?** Ignorer certains éléments sans sortir complètement de la boucle.

## Le `else` des boucles (particularité Python)

Python a une fonctionnalité unique : on peut mettre un `else` après une boucle.

**Ce bloc `else` s'exécute SEULEMENT si la boucle s'est terminée "normalement"** (sans `break`).

````python
nombres = [1, 3, 5, 7, 9]
recherche = 4

for n in nombres:
    if n == recherche:
        print(f'{recherche} trouvé !')
        break
else:
    # Ce bloc s'exécute si on N'A PAS fait break
    print(f'{recherche} non trouvé dans la liste.')
````

````python
# Testons avec un nombre qui existe
nombres = [1, 3, 5, 7, 9]
recherche = 5

for n in nombres:
    if n == recherche:
        print(f'{recherche} trouvé !')
        break
else:
    print(f'{recherche} non trouvé.')  # Ne s'exécute PAS car on a fait break
````

# CHAPITRE 8 : LES FONCTIONS - RÉUTILISER LE CODE
---

## Pourquoi les fonctions existent-elles ?

Imaginez que vous devez calculer la moyenne à plusieurs endroits de votre programme. Sans fonction, vous devriez répéter le même code encore et encore.

**Les fonctions permettent de :**
1. **Réutiliser du code** : Écrire une fois, utiliser plusieurs fois
2. **Organiser le code** : Plus clair et plus lisible
3. **Faciliter la maintenance** : Modifier à un seul endroit

## Qu'est-ce qu'une fonction ?

Une **fonction** est un bloc de code réutilisable qui :
1. Peut recevoir des **entrées** (les paramètres)
2. Fait quelque chose
3. Peut renvoyer un **résultat** (la valeur de retour)

**Analogie :** Une machine dans une usine.
- On met des matières premières (les paramètres)
- La machine les transforme
- On récupère un produit fini (la valeur de retour)

## Créer une fonction

### Structure de base

````python
def dire_bonjour():
    """Cette fonction affiche un message de bienvenue."""
    print('Bonjour tout le monde !')

# Pour utiliser la fonction, on l'appelle :
dire_bonjour()
dire_bonjour()
````

**Éléments :**
1. Le mot-clé `def` (pour "define" = définir)
2. Le nom de la fonction
3. Des parenthèses `()` (même vides)
4. Deux-points `:`
5. Le corps de la fonction, **indenté**

## Les paramètres : donner des informations à la fonction

### Un paramètre

````python
def dire_bonjour(prenom):
    """Affiche un message personnalisé."""
    print(f'Bonjour {prenom} !')

dire_bonjour('Marie')
dire_bonjour('Pierre')
````

### Plusieurs paramètres

````python
def presenter(prenom, age, ville):
    """Présente une personne."""
    print(f'{prenom} a {age} ans et habite à {ville}.')

presenter('Marie', 25, 'Paris')
presenter('Pierre', 30, 'Lyon')
````

### Paramètres avec valeur par défaut

````python
def saluer(prenom, message='Bonjour'):
    """Salue une personne avec un message personnalisable."""
    print(f'{message} {prenom} !')

saluer('Marie')                # Utilise la valeur par défaut
saluer('Marie', 'Bonsoir')     # Utilise la valeur fournie
````

## La valeur de retour : `return`

### À quoi ça sert ?

Souvent, on veut que la fonction **calcule** quelque chose et nous le **renvoie** pour qu'on puisse l'utiliser.

````python
def additionner(a, b):
    """Retourne la somme de deux nombres."""
    resultat = a + b
    return resultat

somme = additionner(5, 3)
print("5 + 3 =", somme)

# On peut utiliser le résultat directement
print("10 + 20 =", additionner(10, 20))
````

### Exemple complet : calculer une moyenne

````python
def calculer_moyenne(notes):
    """Calcule la moyenne d'une liste de notes."""
    if len(notes) == 0:
        return 0
    return sum(notes) / len(notes)

# Utilisation
mes_notes = [12, 15, 14, 16, 13]
moyenne = calculer_moyenne(mes_notes)
print(f'Moyenne : {moyenne:.2f}')
````

### Renvoyer plusieurs valeurs

````python
def analyser_nombre(n):
    """Retourne le carré et le cube d'un nombre."""
    carre = n ** 2
    cube = n ** 3
    return carre, cube

# Récupérer les deux valeurs
carre, cube = analyser_nombre(3)
print(f"3² = {carre}, 3³ = {cube}")
````

## La portée des variables

### Variables locales

Une variable créée **dans** une fonction n'existe que **dans** cette fonction :

````python
def ma_fonction():
    x = 10                  # x est LOCAL à ma_fonction
    print("Dans la fonction: x =", x)

ma_fonction()
# print(x)  # ERREUR ! x n'existe pas ici
````

### Variables globales

````python
x = 10                      # x est GLOBAL

def ma_fonction():
    print("Dans la fonction: x =", x)  # On peut LIRE une variable globale

ma_fonction()
print("Hors de la fonction: x =", x)
````

# CHAPITRE 9 : LA PROGRAMMATION ORIENTÉE OBJET (POO)
---

## Pourquoi la POO existe-t-elle ?

Jusqu'ici, nous avons séparé les **données** (variables) et les **actions** (fonctions). Mais dans la vraie vie, les choses sont différentes :

- Une **voiture** a des caractéristiques (couleur, marque, vitesse) ET peut faire des actions (démarrer, accélérer, freiner)
- Un **compte bancaire** a un solde ET peut recevoir des dépôts et des retraits

La **Programmation Orientée Objet** permet de regrouper les données et les actions dans une même entité appelée **objet**.

## Les concepts clés

### La classe : le "plan de construction"

Une **classe** est comme un plan de construction ou un moule. Elle définit :
- Quelles **caractéristiques** auront les objets (les **attributs**)
- Quelles **actions** pourront faire les objets (les **méthodes**)

### L'objet : une "instance" de la classe

Un **objet** est créé à partir d'une classe. On dit qu'il est une **instance** de la classe.

## Créer une classe

````python
class Voiture:
    """Représente une voiture."""
    
    def __init__(self, marque, couleur):
        """Constructeur : initialise les attributs."""
        self.marque = marque
        self.couleur = couleur
        self.vitesse = 0
    
    def accelerer(self, augmentation):
        """Augmente la vitesse."""
        self.vitesse += augmentation
        print(f'La {self.marque} roule à {self.vitesse} km/h')
    
    def freiner(self, reduction):
        """Réduit la vitesse."""
        self.vitesse -= reduction
        if self.vitesse < 0:
            self.vitesse = 0
        print(f'La {self.marque} roule à {self.vitesse} km/h')

# Créer des objets (instances)
ma_voiture = Voiture('Renault', 'rouge')
ta_voiture = Voiture('Peugeot', 'bleu')

# Utiliser les objets
print(f"Ma voiture est une {ma_voiture.marque} {ma_voiture.couleur}")
ma_voiture.accelerer(50)
ma_voiture.accelerer(30)
ma_voiture.freiner(20)
````

### Explication de `self`

`self` représente **l'objet lui-même**. C'est une référence à l'instance en cours de création ou d'utilisation.

Quand vous écrivez `self.marque = marque`, vous dites : "L'attribut `marque` de **cet objet** prend la valeur du paramètre `marque`."

`self` est TOUJOURS le premier paramètre des méthodes, mais Python le remplit automatiquement quand vous appelez la méthode.

## L'héritage

````python
class Animal:
    """Classe de base pour les animaux."""
    
    def __init__(self, nom):
        self.nom = nom
    
    def manger(self):
        print(f'{self.nom} mange.')


class Chien(Animal):    # Chien HÉRITE de Animal
    """Représente un chien."""
    
    def aboyer(self):
        print(f'{self.nom} fait Wouf !')


class Chat(Animal):     # Chat HÉRITE de Animal
    """Représente un chat."""
    
    def miauler(self):
        print(f'{self.nom} fait Miaou !')


# Utilisation
rex = Chien('Rex')
rex.manger()            # Méthode héritée de Animal
rex.aboyer()            # Méthode propre à Chien

print()

minou = Chat('Minou')
minou.manger()          # Méthode héritée de Animal
minou.miauler()         # Méthode propre à Chat
````

## Pourquoi les exceptions existent-elles ?

Quand quelque chose se passe mal, Python génère une **exception** et arrête le programme. Mais parfois, on veut **anticiper** les erreurs et les **gérer**.

````python
# Sans gestion d'exception - décommentez pour voir l'erreur
# resultat = 10 / 0  # ZeroDivisionError
````

## Gérer les exceptions avec `try-except`

````python
try:
    nombre = int(input('Entrez un nombre : '))
    resultat = 100 / nombre
    print(f'100 / {nombre} = {resultat}')
except ValueError:
    print('Erreur : vous n\'avez pas entré un nombre valide.')
except ZeroDivisionError:
    print('Erreur : division par zéro impossible.')
````
